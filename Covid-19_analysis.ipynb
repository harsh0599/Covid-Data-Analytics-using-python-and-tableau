in which we used pivot table thats help me to check SUM of cured and death cases in state/unionterritory wise 


```python
statewise=pd.pivot_table(covid_df,values=["Confirmed","Deaths","Cured"],
index="State/UnionTerritory",aggfunc=max)
```

Recovery rate is nothing but just to check to find the recover people

Recovery rate = Cured case *100 / confirmed case

```python
statewise["Recovery Rate"]=statewise["Cured"]*100/statewise["Confirmed"]
```

Mortality rate is nothing but just to check to find the recover people

Mortality rate = Death case *100 / confirmed case

```python
statewise["Mortality Rate"]=statewise["Deaths"]*100/statewise["Confirmed"]
```

In which we used the sort function thats help me show the value which we want in asc and dsc order..


```python
statewise=statewise.sort_values(by="Confirmed",ascending=False)
```

# Top 10 Active cases states


In which we used group by function thats is import func in pandas that would help to create the top 10 list of active cases 

And in which we used to defined the value of max values of active cases





```python
top_10_active_cases=covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases','Date']].sort_values(by=['Active_Cases'],ascending= False).reset_index()
```

In which we plot figure with the help of matplotlib function


```python
fig=plt.figure(figsize=(12,8))
```

give title of our asis chart with the help of matplotlibfunc...


```python
plt.title("Top 10 states with most active cases in india",size=25)
```

We uses axis in it which help us to see the graph

make barplot with the help of seaborn library we used Iloc for index location,X and Y axis for for graphs

we use linewidth for gapping btw two blocks and edgecolor for outline the blocks...





```python
ax = sns.barplot(data =top_10_active_cases.iloc[:10], y= "Active_Cases", x= "State/UnionTerritory", linewidth=2, edgecolor='red')
```

In which we use labels thats helps me to remove the overlapping states names this function we used in matplotlib


```python
top_10_active_cases=covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases','Date']].sort_values(by=['Active_Cases'],ascending= False).reset_index()
fig=plt.figure(figsize=(12,8))
plt.title("Top 10 states with most active cases in india",size=25)
ax = sns.barplot(data =top_10_active_cases.iloc[:10], y= "Active_Cases", x= "State/UnionTerritory", linewidth=2, edgecolor='red')
plt.xlabel("states")
plt.ylabel("Total Active Cases")
plt.show()
```

Top 10 death cases in states


```python
top_10_death_case=covid_df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()
fig=plt.figure(figsize=(18,5))
plt.title(" Top 10 states with most deaths",size = 50)
ax=sns.barplot(data=top_10_death_case.iloc[:12],y="Deaths",x="State/UnionTerritory",linewidth=2,edgecolor="red")
plt.xlabel("States")
plt.ylabel("Deaths")
plt.show()
```

we can add new varrable vaccine and remove rows where state = india..


```python
vaccine=vaccine_df[vaccine_df.State!="India"]
vaccine
```

```python
vaccine.rename(columns = {'Total Individuals Vaccinated':'Total'},inplace= True)
vaccine.head()
```

```python
vaccine.rename(columns = {'Total Vaccinated':'Total'},inplace= True)

vaccine.head()
```

most vaccinated state in India

now we would find most vaccinated state in India

we create another variable max_vac

```python
max_vac=vaccine.groupby('State')['Total'].sum().to_frame('Total')
```

```python
max_vac=max_vac.sort_values('Total',ascending=False)[:5]
max_vac
```

Data converted into charts

```python
fig=plt.figure(figsize=(10,5))
plt.title("Top 5 vaccinated states in India",size=20)
x=sns.barplot(data=max_vac.iloc[:10],y=max_vac.Total,x=max_vac.index,linewidth=2,edgecolor='Black')
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```

least vaccination states in india

```python
max_vac=vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac=max_vac.sort_values('Total',ascending=False)[-6:-1]
max_vac
```

```python
fig=plt.figure(figsize=(10,5))
plt.title("Least 5 vaccinated states in India",size=20)
x=sns.barplot(data=max_vac.iloc[:10],y=max_vac.Total,x=max_vac.index,linewidth=2,edgecolor='Black')
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
